# **F.) Object-Oriented Programming**

- OOD helps us organize our computer code to solve problems easily.
- We think of our code as LEGO bricks.
- We create objects that represent things in the real world, like person or car.
- Objects have properties (like name, age ..).
- We can make new objects based on existing ones, adding extra features if needed. (like teacher and student)

```python
display.Image("img/ood.png",width = 800, height = 100)
```

```python
class Person:
    def __init__(self, name, age):
        self.name = name
        self.age = age
    def __repr__(self):
        return f"Person(name={self.name}, age={self.age})"
    def print_details(self):
        print(f"Name: {self.name}, Age: {self.age}")


class Teacher(Person):
    def __init__(self, name, age, subject):
        super().__init__(name, age)
        self.subject = subject
    def print_details(self):
        super().print_details()
        print(f"Subject: {self.subject}")

class Student(Person):
    def __init__(self, name, age, net_id):
        super().__init__(name, age)
        self.net_id = net_id
    def print_details(self):
        super().print_details()
        print(f"Student ID: {self.net_id}")


# Example usage
person = Person("Alice", 30)
teacher = Teacher("Frances", 25, "Python")
student = Student("Bob", 20, "S12345")

# Print details
person.print_details()
teacher.print_details()
student.print_details()

```
